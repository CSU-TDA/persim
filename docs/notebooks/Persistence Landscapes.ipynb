# Persistence Landscapes
---

This notebook gives usage examples for the persistence landscape functionality in `persim`. We will construct persistence landscapes for two different datasets (a sphere and a torus), plot them in comparison with each other, plot average persistence landscapes, and plot differences of average landscapes.

```python
# imports

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from ripser import Rips
from tadasets import torus, sphere
import persim
```

```python
data_torus = torus(n=100, c=2, a=2, ambient=10, noise=0.2)
data_sphere = sphere(n=100, r=2, ambient=10, noise=0.2)
```

```python
rips = Rips(maxdim = 2)
```

```python
dgms_torus = rips.fit_transform(data_torus)
dgms_sphere = rips.fit_transform(data_sphere)
```

```python
# Visualize the data
fig, axs = plt.subplots(1, 2)
fig.set_size_inches(10, 5)

persim.plot_diagrams(dgms_torus, title="Persistence Diagram of Torus", ax=axs[0])

persim.plot_diagrams(dgms_sphere, title="Persistence Diagram of Sphere", ax=axs[1])

fig.tight_layout()
```

```python
landscapes_torus = []
landscapes_sphere = []

for d in range(rips.maxdim + 1):
    landscapes_torus.append(persim.PersLandscapeExact(dgms_torus, hom_deg=d))
    
    landscapes_sphere.append(persim.PersLandscapeExact(dgms_sphere, hom_deg=d))
    
```

```python
#TODO We should pass axes into the basic plotting functions so we can do cool splits like above
#TODO We should also fix the legend size for lower degrees and larger landscapes

# Visualize the data
fig, axs = plt.subplots(1, 2)
fig.set_size_inches(10, 5)

persim.plot_landscape_simple(landscapes_torus[2],
                             title="Degree 2 Persistence Landscape of Torus", ax=axs[0])

persim.plot_landscape_simple(landscapes_sphere[2],
                            title="Degree 2 Persistence Landscape of Sphere", ax=axs[1])

fig.tight_layout()
```
